```python
# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
### show the result of using a right join on the DataFrames
### changed the how parameter to 'right'

result = users.merge(roles, left_on='role_id', right_on='id', how='right', indicator=True)
result
```

```python
### show the result of using an outer join on the DataFrames 
### change the how parameter to 'outer'.

result = users.merge(roles, left_on='role_id', right_on='id', how='outer', indicator=True)
result
```

> If we drop the foreign keys from the DataFrames and try to merge them, we will get an error. The error stems from the merge operation which requires a common column to join the DataFrames on. If the common colums is dropped, the data cannot be merged correctly.

```python
### load the mpg dataset from PyDataset and output and read the documentation

from pydataset import data

mpg = data('mpg')
print(data('mpg', show_doc=True))
```

```python
### Show the number of rows and columns in the dataset, use the shape attribute.

print('Number of rows and columns in the dataset:', mpg.shape)
```

```python
### Check column names and perform industry standard cleanup, use the columns attribute to get a list of column names and the rename() method to rename columns as needed.

print('Column names before cleanup:\n', mpg.columns)

mpg = mpg.rename(columns={'cty': 'city', 'hwy': 'highway','displ': 'displacement', 'class': 'vehicle_class'})
print('Column names after cleanup:\n', mpg.columns)
```

```python
mpg
```

```python
### display the summary statistics for the dataset, use the describe() method.

mpg.describe()
```

```python
### print how many different manufacturers are in the dataset, 
### use the nunique() method on the 'manufacturer' column.

print('Number of different manufacturers:', mpg['manufacturer'].nunique())
```

```python
### print how many different models are in the dataset, 
### use the nunique() method on the 'model' column.

print('Number of different models:', mpg['model'].nunique())
```

```python
### create a column named mileage_difference, 
### subtract the 'city' column from the 'highway' column.

mpg['mileage_difference'] = mpg['highway'] - mpg['city']
```

```python
mpg
```

```python
### create a column named average_mileage, 
### take the mean of the 'city' and 'highway' columns.

mpg['average_mileage'] = (mpg['city'] + mpg['highway']) / 2
```

```python
mpg
```

```python
### create a new column named 'is_automatic' that holds boolean values 
### denoting whether the car has an automatic transmission or not

mpg['is_automatic'] = mpg['trans'].str.startswith('auto')
```

```python
### find out which manufacturer has the best miles per gallon on average,
# group the data by manufacturer, 
# compute the mean of the 'hwy' column, and sort the result in descending order.

manu_top_miles = mpg.groupby('manufacturer')['average_mileage'].mean().sort_values(ascending=False)
```

```python
# removes the dollar sign from the item_price column and converts it to a float.
orders['item_price'] = orders['item_price'].str.replace('$', '').astype(float)
```

```python
# total price for the order
order_totals = orders.groupby('order_id')['item_price'].sum()
```

```python
### order totals by item name
order_items = orders.groupby('item_name')['quantity'].sum()
```

```python
order_items
```

```python
### find the most popular 3 items
### group the orders by item_name and sum the quantity for each item. 
### use the nlargest method to find the top 3 items.
popular_items = orders.groupby('item_name')['quantity'].sum().nlargest(3)
print(popular_items)
```

```python
### item that produced the most revenue
item_revenue = orders.groupby('item_name')['item_price'].sum().nlargest(1)
print(item_revenue)
```
