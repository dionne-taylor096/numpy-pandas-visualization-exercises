> Using the chipotle dataset from the previous exercise, 
> create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
#Load data from csv
orders = pd.read_csv('orders.csv')
```

```python
orders.head(10)
```

```python
orders.info()
```

```python
#removes the dollar sign from the item_price column and converts it to a float.
orders['item_price'] = orders['item_price'].str.replace('$', '').astype(float)
```

```python
#order totals by item name
order_items = orders.groupby('item_name')['quantity'].sum()
```

```python
order_items
```

```python
# find the most popular 3 items
# group the orders by item_name and sum the quantity for each item. 
# use the nlargest method to find the top 3 items.
popular_items = orders.groupby('item_name')['quantity'].sum().nlargest(4)
```

```python
popular_items
```

```python
# items that produced the most revenue
item_revenue = orders.groupby('item_name')['item_price'].sum().nlargest(4)
```

```python
item_revenue
```

```python
# Group the orders by item name and calculate the popularity and revenue for each item
item_data = orders.groupby('item_name').agg({'quantity': 'sum', 'item_price': 'sum'})
```

```python
item_data
```

```python
# Sort the items by quantity in descending order and take the top 4
popular_items = item_data.sort_values(by='quantity', ascending=False).head(4)
```

```python
popular_items
```

```python
# Create a bar chart showing the revenue for the top 4 most popular items
sns.barplot(y="item_price", x=popular_items.index, data=popular_items, palette="Blues_d")
plt.title("Top 4 Chipotle Popular Items and Their Revenue")
plt.xlabel("Revenue")
plt.ylabel("Item Name")
plt.show()
```
