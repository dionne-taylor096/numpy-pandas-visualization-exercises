```python
# Create data

# Set time values in increments of 20 
time = range(1, 600, 20)

## Set initial number of bacteria equal to 1
initial = 1


bacteria = [ initial * np.float_power(2, (0.05*n)) for n in time ]
```

```python
## Plot the values
plt.plot(time, bacteria, marker = "o")
plt.title('E.coli grows exponentially faster', fontsize = 20)
plt.xlabel('Time(minutes)')
plt.ylabel('Number of Bacteria')
```

```python
x = range(-4, 5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize=(14, 6)) # (width, height)

plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')
plt.plot(x, sin_x, c='blue', label='$\sin(x)$')
plt.plot(x, cos_x, c='orange', label='$\cos(x)$')

plt.legend(loc='upper right')
plt.title('Approximate sin and cosine curves')
plt.xlabel('$x$')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x points
x = np.linspace(-5, 5, 100)

# Compute y points
y = x**2 - x + 2

# Create the graph
plt.plot(x, y)

# Add annotation for the point (0,0)
plt.annotate("Origin", xy=(0,0), xytext=(1,10),
             arrowprops=dict(facecolor='black', shrink=0.05))
             
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = x^2 - x + 2')

# Display the graph
plt.show()
```

```python
# Equation 1: y = sqrt(x)
x = np.linspace(0, 10, 100)
y = np.sqrt(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = sqrt(x)')

# Equation 2: y = x^3
x = np.linspace(-5, 5, 100)
y = x**3
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = x^3')

# Equation 3: y = 2x
x = np.linspace(-5, 5, 100)
y = 2*x
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = 2x')

# Equation 4: y = 1/(x+1)
x = np.linspace(-10, 10, 100)
y = [1/(i+1) for i in x]
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = 1/(x+1)')

# Display the graphs
plt.show()
```

```python
# Create the figure with 4 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Equation 1: y = sqrt(x)
x = np.linspace(0, 10, 100)
y = np.sqrt(x)
axs[0, 0].plot(x, y)
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].set_title('y = sqrt(x)')

# Equation 2: y = x^3
x = np.linspace(-5, 5, 100)
y = x**3
axs[0, 1].plot(x, y)
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].set_title('y = x^3')
                    
# Equation 3: y = 2x

x = np.linspace(-5, 5, 100)
y = 2*x
axs[1, 0].plot(x, y)
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].set_title('y = 2x')

# Equation 4: y = 1/(x+1)

x = np.linspace(-10, 10, 100)
y = [1/(i+1) for i in x]
axs[1, 1].plot(x, y)
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].set_title('y = 1/(x+1)');
```

```python
# Create the figure with 4 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
# Add a common title and legend for the figure
fig.suptitle('Graphs of 4 Equations')
# Adjust spacing between subplots
fig.subplots_adjust(hspace=0.4, wspace=0.3)

# Equation 1: y = sqrt(x)
x = np.linspace(0, 10, 100)
y = np.sqrt(x)
axs[0, 0].plot(x, y)
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].set_title('y = sqrt(x)')

# Equation 2: y = x^3
x = np.linspace(-5, 5, 100)
y = x**3
axs[0, 1].plot(x, y)
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].set_title('y = x^3')
                    
# Equation 3: y = 2x

x = np.linspace(-5, 5, 100)
y = 2*x
axs[1, 0].plot(x, y)
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].set_title('y = 2x')

# Equation 4: y = 1/(x+1)

x = np.linspace(-10, 10, 100)
y = [1/(i+1) for i in x]
axs[1, 1].plot(x, y)
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].set_title('y = 1/(x+1)');
```

```python
# Create the figure with 4 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
# Add a common title and legend for the figure
fig.suptitle('Graphs of 4 Equations')

# Adjust spacing between subplots
fig.subplots_adjust(hspace=0.4, wspace=0.3)


## Added colors to the subplots
# Equation 1: y = sqrt(x)
x = np.linspace(0, 10, 100)
y = np.sqrt(x)
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].set_title('y = sqrt(x)')

# Equation 2: y = x^3
x = np.linspace(-5, 5, 100)
y = x**3
axs[0, 1].plot(x, y, color='red')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].set_title('y = x^3')

# Equation 3: y = 2x
x = np.linspace(-5, 5, 100)
y = 2*x
axs[1, 0].plot(x, y, color='green')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].set_title('y = 2x')

# Equation 4: y = 1/(x+1)
x = np.linspace(-10, 10, 100)
y = [1/(i+1) for i in x]
axs[1, 1].plot(x, y, color='purple')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].set_title('y = 1/(x+1)');
```
